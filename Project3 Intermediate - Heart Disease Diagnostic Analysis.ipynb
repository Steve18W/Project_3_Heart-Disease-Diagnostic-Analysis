# Heart Disease Diagnostic Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv("Heart Disease Data.csv")
Data

```

```python
print(Data.describe())

```

```python

print("\nMissing values in the dataset:")
print(Data.isnull().sum())

```

```python
print("\nSummary Statistics:")
print(Data.describe())

```

```python
#Generating a Correlation Matrix Heatmap for Visualizing Feature Relationships

correlation_matrix = Data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

```

```python
#Visualizing Heart Disease Distribution by Gender Using a Count Plot

plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=Data, hue='target', palette='viridis')
plt.title('Heart Disease Distribution by Gender')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.legend(['No Disease', 'Disease'])
plt.show()


```

```python
#Stacked Histogram of Heart Disease Rates by Age

plt.figure(figsize=(12, 6))
sns.histplot(x='age', data=Data, hue='target', multiple='stack', bins=20, palette='viridis')
plt.title('Heart Disease Rates by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()


```

```python
# Scatter plot of age vs target


plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='thalach', hue='target', data=Data, palette='coolwarm')
plt.title('Max Heart Rate (Thalach) vs Age')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate (Thalach)')
plt.show()
```
